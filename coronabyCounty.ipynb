```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from vega_datasets import data
```

```python
# loading data
corona = pd.read_csv('~/Desktop/Code n Stuff/pghcoffee.csv')
columnsList = list(corona.columns)

# preparing data
corona_spring = corona.sort_values(['State', 'County Name'])
corona_spring.reset_index(drop=True, inplace=True)
corona_spring
```

```python
# separate data into 51 separate datasets
def filterByState():
    states = list(corona_spring.State.unique())
    numOfCounties = len(corona_spring)
    byStates = []

    for state in states:
        corona_byState = corona_spring[corona_spring['State'] == state]
        byStates.append(corona_byState)
    return byStates

# transpose the 51 datasets 
def transposeByDate():
    byDates = []
    
    corona_byStates = filterByState()
    for item in corona_byStates:
        byDate = item.transpose()
        byDate_Mar1 = byDate.loc['3/10/2020':,]
        byDates.append(byDate)
    return byDates

corona_byDates = transposeByDate()

# plotting all 51 datasets
print(type(corona_byDates[0]))
graph = corona_byDates[0].plot(kind='line')
graph.show
#for item in corona_byDates:
    #print(item)
    #graph = item.plot(kind = 'line', legend=None)
    #graph.show()
```
